```python
#q-learning with 2 agv
import import_ipynb
from maze_env import Maze
#from RL_brain import QLearningTable
from dqn import DQN
import time
import simpy
import random
import numpy as np


class AGV:
    def __init__(self,idx,begin,destination,state=0,acc_req = None,move_req=0):
        self.idx=idx
        self.last=begin
        self.begin=begin
        self.des=destination
        self.cur=begin
        self.state=state
        self.acc_req = acc_req
        self.move_req = move_req
        
class Request:
    def __init__(self,stocker,req_set,start=None,end=None):
        self.stocker=stocker
        self.req_set=req_set
        self.deli_start=start
        self.deli_end=end
        self.des=self.decide_des()
        #self.des=self.decide()
        
    def decide(self):
        if self.stocker==0:
            return 2
        elif self.stocker==1:
            return 3
        elif self.stocker==2:
            return 0
        elif self.stocker==3:
            return 1
        
    def decide_des(self):
        while True:
            des = np.random.choice(len(self.req_set))
            if des != self.stocker:
                return des

class node:
    def __init__(self,occupy=0):
        self.occupy = occupy
        self.next = 0
        self.belong=None
        
def move_where(action,node_ind):
    if action==0:
        if node_ind==3 or node_ind==5:
            base_action = [UNIT, -UNIT]
        elif node_ind==8 or node_ind==16:
            base_action = [0, -UNIT]
        elif node_ind==12 or node_ind==13:
            base_action = [0, -2*UNIT]
        elif node_ind==22 or node_ind==23:
            base_action = [-UNIT, -UNIT]

    elif action==1:
        if node_ind==0 or node_ind==1:
            base_action = [UNIT, UNIT]
        elif node_ind==7 or node_ind==15:
            base_action = [0, UNIT]
        elif node_ind==10:
            base_action = [0, 2*UNIT]
        elif node_ind==18 or node_ind==20:
            base_action = [-UNIT, UNIT]

    elif action==2:
        if node_ind==2 or node_ind==4 or node_ind==6:
            base_action = [UNIT, 0]
        elif node_ind==3 or node_ind==5 or node_ind==7 or node_ind==12 or node_ind==14 or node_ind==15:
            base_action = [2*UNIT, 0]
        elif node_ind==13:
            base_action = [3*UNIT, 0]

    elif action==3:
        if node_ind==17 or node_ind==19 or node_ind==21:
            base_action = [-UNIT, 0]
        elif node_ind==9 or node_ind==11 or node_ind==18 or node_ind==20:
            base_action = [-2*UNIT, 0]
        elif node_ind==10:
            base_action = [-3*UNIT, 0]
    return  base_action
        
def front_agv_recur(now_node,node_set,record,direction,unit_len,have_record):
    unit_len+=1
    now_posi = grid[now_node]
    going_record={}
    ava_action = action_space[now_node]
    for action in ava_action:
        base_action = move_where(action,now_node)
        next_posi = np.sum([now_posi,base_action],axis=0).tolist()
        next_node = list(grid.keys())[list(grid.values()).index(next_posi)]
        if next_node not in have_record:
            going_record[action] = next_node
            
    have_record.extend(list(going_record.values()))
    for action,next_node in list(going_record.items()):
        #if node_set[next_node].occupy==1:
        occupied=node_set[next_node].belong
        if occupied!=None:
            record[direction]+=1
        if unit_len<2:
            front_agv_recur(next_node,node_set,record,direction,unit_len,have_record)
    
def front_agv(agv,node_set):
    record = [-1]*4
    have_record=[]
    going_record={}
    unit_len=0
    now_node = list(grid.keys())[list(grid.values()).index(agv.cur)]
    ava_action = action_space[now_node]
    for action in ava_action:
        base_action = move_where(action,now_node)
        next_posi = np.sum([agv.cur,base_action],axis=0).tolist()
        next_node = list(grid.keys())[list(grid.values()).index(next_posi)]
        going_record[action]=next_node
        record[action]=0
    
    have_record.extend(list(going_record.values()))
    for action,next_node in list(going_record.items()):
        #if node_set[next_node].occupy==1:
        occupied=node_set[next_node].belong
        if occupied!=None:
            record[action]+=1
        front_agv_recur(next_node,node_set,record,action,unit_len,have_record)
    return record
        
def define_request(stocker,sim,req_set):
    print("stocker %d request %d generate at %.2f" % (stocker, len(req_set[stocker]),sim.now))
    
def initial_request(sim,req_set,stocker):
    define_request(stocker,sim,req_set)

def generate_request(sim,req_set,req_cor,stocker):
    while True:
        yield sim.timeout(REQ_TIME)
        req = Request(stocker,req_set)
        req_set[stocker].append(req)
        define_request(stocker,sim,req_set)
        env.update_req(req_set,req_cor,stocker)
        

def generate_req(sim,req_set,stocker):
    req = Request(stocker,req_set)
    req_set[stocker].append(req)
    initial_request(sim,req_set,stocker)
    
    if stocker==0:
        req_cor=[220,60]
    elif stocker==1:
        req_cor=[340,60]
    elif stocker==2:
        req_cor=[220,380]
    elif stocker==3:
        req_cor=[340,380]
    
    env.update_req(req_set,req_cor,stocker,init=True)
    sim.process(generate_request(sim,req_set,req_cor,stocker))

def update_state(env,agv,node_set,next_posi,next_node_ind,node_ind):
    agv.last = agv.cur
    agv.cur = next_posi

    print('agv',agv.idx,'cur position',agv.cur,'last position',agv.last)
    #node_set[next_node_ind].occupy = 1
    #node_set[node_ind].occupy = 0
    print('agv',agv.idx,'cur occupy',node_set[next_node_ind].occupy,'last occupy',node_set[node_ind].occupy)
    trans_fini = check_posi(env,agv,req_set)
    return trans_fini
    
def check_posi(env,agv,req_set):
    if agv.state==1:
        if agv.cur==agv.des:
            stocker = list(stocker_pos.keys())[list(stocker_pos.values()).index(agv.cur)]
            print('agv',agv.idx,'traverse arrive at stocker',stocker)
            agv.state=2
            trans_stoc = agv.acc_req
            agv.des = stocker_pos[trans_stoc.des]
            print('agv',agv.idx,'start to transport to stocker',trans_stoc.des)
            
            return True
    
    elif agv.state==2:
        if agv.cur==agv.des:
            acc_req=agv.acc_req
            trans_time=sim.now-acc_req.deli_start
            if sim.now not in list(all_trans_time.keys()):
                all_trans_time[sim.now]=[trans_time]
            else:
                all_trans_time[sim.now].append(trans_time)
            
            unser_req = [len(req_set[i]) for i in range(len(req_set))]
            sto_dict={}
            for i in range(len(unser_req)):
                if unser_req[i]!=0:
                    dist = np.sum((np.array(agv.cur)-np.array(stocker_pos[i]))**2)
                    sto_dict[i]=dist
            #跟state 0 一樣的程式
            if len(sto_dict)!=0:
                next_stoc = list(sto_dict.keys())[list(sto_dict.values()).index(min(list(sto_dict.values())))]
                if stocker_pos[next_stoc]==agv.cur:
                    agv.state=2
                    acc_req=req_set[next_stoc][0]
                    acc_req.deli_start=sim.now
                    agv.acc_req = acc_req
                    agv.des = stocker_pos[acc_req.des]
                    req_set[next_stoc].pop(0)

                    if next_stoc==0:
                        req_cor=[220,60]
                    elif next_stoc==1:
                        req_cor=[340,60]
                    elif next_stoc==2:
                        req_cor=[220,380]
                    elif next_stoc==3:
                        req_cor=[340,380]

                    env.update_req(req_set,req_cor,next_stoc)
            
                else:
                    agv.state = 1
                    agv.des = stocker_pos[next_stoc]
                    acc_req=req_set[next_stoc][0]
                    acc_req.deli_start=sim.now
                    agv.acc_req = acc_req
                    req_set[next_stoc].pop(0)

                    if next_stoc==0:
                        req_cor=[220,60]
                    elif next_stoc==1:
                        req_cor=[340,60]
                    elif next_stoc==2:
                        req_cor=[220,380]
                    elif next_stoc==3:
                        req_cor=[340,380]

                    env.update_req(req_set,req_cor,next_stoc)

                #如果所有倉儲都沒有要求 len(sto_dict)=0 原地等待
            else:
                agv.state = 0
                #print(agv.cur)
                #env.reset_agv(agv)
            #agv.state = 2
            #agv.des = stocker_pos[0]
            return True
        
    elif agv.state==0:
        #find out unservice request
        unser_req = [len(req_set[i]) for i in range(len(req_set))]
        sto_dict={}
        for i in range(len(unser_req)):
            if unser_req[i]!=0:
                dist = np.sum((np.array(agv.cur)-np.array(stocker_pos[i]))**2)
                sto_dict[i]=dist

        if len(sto_dict)!=0:
            next_stoc = list(sto_dict.keys())[list(sto_dict.values()).index(min(list(sto_dict.values())))]
            if stocker_pos[next_stoc]==agv.cur:
                agv.state=2
                acc_req=req_set[next_stoc][0]
                acc_req.deli_start=sim.now
                agv.acc_req = acc_req
                agv.des = stocker_pos[acc_req.des]
                req_set[next_stoc].pop(0)
                
                if next_stoc==0:
                    req_cor=[220,60]
                elif next_stoc==1:
                    req_cor=[340,60]
                elif next_stoc==2:
                    req_cor=[220,380]
                elif next_stoc==3:
                    req_cor=[340,380]
                
                env.update_req(req_set,req_cor,next_stoc)
            
            else:
                agv.state = 1
                agv.des = stocker_pos[next_stoc]
                acc_req=req_set[next_stoc][0]
                acc_req.deli_start=sim.now
                agv.acc_req = acc_req
                req_set[next_stoc].pop(0)

                if next_stoc==0:
                    req_cor=[220,60]
                elif next_stoc==1:
                    req_cor=[340,60]
                elif next_stoc==2:
                    req_cor=[220,380]
                elif next_stoc==3:
                    req_cor=[340,380]

                env.update_req(req_set,req_cor,next_stoc)
            
    return False


def update(sim,env,agv,node_set):
    # initial observation
    #print("agv%d" % (agv.idx))
    #record=[0]*4
    #find out observation
    #record = front_agv(agv,node_set)
    #print(record)
    #observation=str(agv.cur + agv.des)

    while True:
        if agv.state!=0:
            # fresh env
            env.render()
            '''q-learning
            observation=str(agv.cur + agv.des)'''
            record = front_agv(agv,node_set)
            observation=agv.cur + agv.des + record
            # RL choose action based on observation
            node_ind = list(grid.keys())[list(grid.values()).index(agv.cur)]
            print('agv',agv.idx,'at node',node_ind)
            '''q-learning
            action = RL.boltzmann_softmax(node_ind,observation)'''
            action = RL.select_action(node_ind,observation)
            print('agv',agv.idx,'action',action)

            #collision check(if occupy then can't move)
            base_action = move_where(action,node_ind)
            next_posi = np.sum([agv.cur,base_action],axis=0).tolist()
            next_node_ind = list(grid.keys())[list(grid.values()).index(next_posi)]

            before_move = sim.now

            while True:
                print('agv',agv.idx,'next position',next_posi)
                if node_set[next_node_ind].occupy == 0:
                    print('agv',agv.idx,'is occupy?',node_set[next_node_ind].occupy)
                    node_set[next_node_ind].occupy=1
                    node_set[next_node_ind].belong=agv
                    node_set[node_ind].occupy=0
                    node_set[node_ind].belong=None
                    break
                else:                       
                    print('agv%d stop time %d' % (agv.idx,sim.now))
                    agv_pos = [v.cur for v in agv_set]
                    if next_posi in agv_pos:
                        which_agv = agv_pos.index(next_posi)
                        if agv_set[which_agv].state==0:
                            agv_set[which_agv].move_req=1
                            print('agv',agv.idx,'request agv',which_agv,'to move')
                    yield sim.timeout(1)

            # RL take action and get next observation and reward
            yield sim.timeout(1)
            after_move = sim.now
            print('agv',agv.idx,'arrive at',sim.now)
            trans_fini = update_state(env,agv,node_set,next_posi,next_node_ind,node_ind)
            
            ind = list(stocker_pos.keys())[list(stocker_pos.values()).index(agv.des)]
            env.new_step(agv,base_action,trans_fini,ind)
            
            record = front_agv(agv,node_set)
            '''q-learning
            observation_ = str(agv.cur + agv.des)'''
            observation_ = agv.cur + agv.des + record
            reward = after_move - before_move
            # RL learn from this transition
            '''q-learning
            RL.learn(observation, action, -reward, observation_, trans_fini,next_node_ind,sim.now)'''
            RL.store(observation, action, reward, observation_, trans_fini)
            RL.learn(grid)

            # swap observation
            '''q-learning
            observation = observation_'''
        else:
            yield sim.timeout(1)
            trans_fini = check_posi(env,agv,req_set)
            #
            if agv.state!=0:
                print('agv',agv.idx,'find out request at',sim.now)
                ind = list(stocker_pos.keys())[list(stocker_pos.values()).index(agv.des)]
                env.reset_agv(agv,ind)
                agv.move_req=0
            else:
                if agv.move_req==1:
                    idle_node_ind = list(grid.keys())[list(grid.values()).index(agv.cur)]
                    ava_action = action_space[idle_node_ind]
                    is_occupy={}
                    for ac in ava_action:
                        idle_base_action = move_where(ac,idle_node_ind)
                        idle_next_posi = np.sum([agv.cur,idle_base_action],axis=0).tolist()
                        idle_next_node_ind = list(grid.keys())[list(grid.values()).index(idle_next_posi)]
                        if node_set[idle_next_node_ind].occupy==0:
                            is_occupy[ac]=0
                        else:
                            tra_agv = node_set[idle_next_node_ind].belong
                            if tra_agv.cur==idle_next_posi and  tra_agv.state==0:
                                is_occupy[ac]=1
                    
                    if len(is_occupy)!=0:
                        if sum(list(is_occupy.values()))<len(ava_action):
                            if sum(list(is_occupy.values()))==0:
                                #從values中選擇 不要直接從ava_action中選 因為有些動作根本沒被加進來
                                #ran_action = np.random.choice(ava_action)
                                ran_action = np.random.choice(list(is_occupy.keys()))
                            else:
                                #建立在動作最多只有2個的情況下
                                if 0 in list(is_occupy.values()):
                                    ran_action = list(is_occupy.keys())[list(is_occupy.values()).index(0)]
                                else:
                                    ran_action = list(is_occupy.keys())[0]
                        else:
                            ran_action = np.random.choice(list(is_occupy.keys()))
                            
                        idle_base_action = move_where(ran_action,idle_node_ind)
                        idle_next_posi = np.sum([agv.cur,idle_base_action],axis=0).tolist()
                        idle_next_node_ind = list(grid.keys())[list(grid.values()).index(idle_next_posi)]
                        
                        
                        #for roun in range(3):
                        while True:
                            if node_set[idle_next_node_ind].occupy == 0:
                                #print('agv',agv.idx,'is occupy?',node_set[next_node_ind].occupy)
                                node_set[idle_next_node_ind].occupy=1
                                node_set[idle_node_ind].occupy=0
                                node_set[idle_next_node_ind].belong=agv
                                node_set[idle_node_ind].belong=None
                                print('idle agv',agv.idx,'are requested to move to',idle_next_node_ind,'at',sim.now)
                                break
                            else:                       
                                print('idle agv%d stop time %d' % (agv.idx,sim.now))
                                agv_pos = [v.cur for v in agv_set]
                                print(agv_pos)
                                which_agv = agv_pos.index(idle_next_posi)
                                agv_set[which_agv].move_req=1
                                print('idle agv',agv.idx,'request idle agv',which_agv,'to move')
                                yield sim.timeout(1)
                                
                        yield sim.timeout(1)
                        agv.last = agv.cur
                        agv.cur = idle_next_posi
                        print('idle agv',agv.idx,'arrived at',sim.now)
                        print('idle agv',agv.idx,'cur position',agv.cur,'last position',agv.last)
                        print('idle agv',agv.idx,'cur occupy',node_set[idle_next_node_ind].occupy,'last occupy',node_set[idle_next_node_ind].occupy)
                        env.new_step(agv,idle_base_action,trans_fini=False,ind='N')
                        agv.move_req=0
            #yield sim.timeout(1)
            
            

def main(sim,env,req_set,agv_set,node_set):
    #generate request at each stocker
    for req_num in range(len(req_set)):
        generate_req(sim,req_set,stocker=req_num)
    yield sim.timeout(1)
    time.sleep(1)

    #initiate agv at stocker
    ini=[0,1,2,3]
    ran=list(np.random.choice(range(2,22),AGV_NUM-len(ini),replace=False))
    al = ini+ran
    for agv_num in range(len(al)):
        print(f'agv {agv_num} generate at {sim.now}')
        if agv_num<4:
            if agv_num==0:
                begin=[220,100]
                destination=[220,100]
            elif agv_num==1:
                begin=[340,100]
                destination=[340,100]
            elif agv_num==2:
                begin=[220,340]
                destination=[220,340]
            elif agv_num==3:
                begin=[340,340]
                destination=[340,340]
        else:
            begin=grid[al[agv_num]]
            destination=[0,0]
        
        print(f'agv {agv_num} generate at {begin}')
        agv=AGV(agv_num,begin,destination,state=0)
        agv_set.append(agv)
        #check function, let agv.state=2 and agv
        '''trans_fini = check_posi(env,agv,req_set)
        ind = list(stocker_pos.keys())[list(stocker_pos.values()).index(agv.des)]'''
        n_ind = list(grid.keys())[list(grid.values()).index(agv.begin)]
        node_set[n_ind].occupy=1
        node_set[n_ind].belong=agv
        env.generate_agv(agv,begin)

        #agv start move
        sim.process(update(sim,env,agv,node_set))
        yield sim.timeout(0)     
                        

UNIT = 40
REQ_NUM=4
NODE_NUM=24
REQ_TIME=5
AGV_NUM=4
#目前的node除以24的商 乘上某倍數來得到座標
grid={0:[220,100],1:[340,100],2:[140,140],3:[180,140],4:[260,140],5:[300,140],
     6:[380,140],7:[420,140],8:[140,180],9:[220,180],10:[340,180],11:[420,180],
     12:[140,260],13:[220,260],14:[340,260],15:[420,260],16:[140,300],17:[180,300],
     18:[260,300],19:[300,300],20:[380,300],21:[420,300],22:[220,340],23:[340,340]}

stocker_pos={0:[220,100],1:[340,100],2:[220,340],3:[340,340]}

action_space = {0:[1],1:[1],2:[2],3:[0,2],4:[2],5:[0,2],6:[2],7:[1],8:[0],9:[3],10:[1,3],11:[3],
                12:[0,2],13:[0,2],14:[2],15:[1],16:[0],17:[3],18:[1,3],19:[3],20:[1,3],21:[3],22:[0],23:[0]}

req_set = [[]*i for i in range(REQ_NUM)]
agv_set = []
node_set = [node() for i in range(NODE_NUM)]
all_trans_time={}
'''q-learning
RL = QLearningTable(actions=list(range(4)))'''
RL = DQN()
env = Maze()
sim = simpy.Environment()
sim.process(main(sim,env,req_set,agv_set,node_set))
sim.run(until=2000)
env.mainloop()
```
