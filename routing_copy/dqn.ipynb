```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from collections import deque
import random

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        self.fc1 = nn.Linear(8, 16)
        #self.fc2 = nn.Linear(16, 16)
        self.fc3 = nn.Linear(16, 4)

    def forward(self, x):
        x = F.leaky_relu(self.fc1(x))
        #x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
class DQN:
    def __init__(self,):
        self.evaluate_net = Net()
        self.target_net = Net()
        self.lr = 0.0025
        self.optimizer = torch.optim.RMSprop(self.evaluate_net.parameters(), lr=self.lr)#, alpha=0.95, eps=0.01
        self.loss_func = torch.nn.MSELoss()
        self.loss_func = self.loss_func.cuda()
        self.evaluate_net = self.evaluate_net.cuda()
        self.target_net = self.target_net.cuda()
        self.learn_step_counter = 0
        self.batch_size = 64
        self.gamma = 0.9
        self.target_replace = 50
        self.learn_frequent = 5
        self.learn_start = 500
        self.decay = 0.9
        self.t = 1
        self.t_min = 0.01
        self.update_times = 0
        self.memory_size = 2000
        self.replay_memory = deque(maxlen=self.memory_size)
        self.action_space = {0:[1],1:[1],2:[2],3:[0,2],4:[2],5:[0,2],6:[2],7:[1],8:[0],9:[3],10:[1,3],11:[3],
                            12:[0,2],13:[0,2],14:[2],15:[1],16:[0],17:[3],18:[1,3],19:[3],20:[1,3],21:[3],22:[0],23:[0]}
        
    def select_action(self, node_ind, s):
        s = torch.cuda.FloatTensor(s)
        q_value = self.evaluate_net(s).detach().cpu().data.numpy()
        '''q_value = q_value[self.action_space[node_ind]]
        
        q_avg = np.average(q_value)

        if q_avg != 0:
            q_value = q_value/q_avg

        proba = np.exp(-q_value/self.t)/sum(np.exp(-q_value/self.t))

        ran = np.random.uniform()
        if ran <= proba[0]:
            action = self.action_space[node_ind][0]
        else:
            action = self.action_space[node_ind][1]'''
        
        if np.random.uniform() <= self.t:
            de_action = np.random.choice(self.action_space[node_ind])
        else:
            while True:
                de_action = np.random.choice(np.where(q_value==q_value[self.action_space[node_ind]].min())[0])
                if de_action in self.action_space[node_ind]:
                    break
        #print(de_action)
        return de_action
    
    def store(self, s, a, r, s_, trans_fini):
        total=s
        total.append(a)
        total.append(r)
        total.extend(s_)
        total.append(trans_fini)
        self.replay_memory.append(total)
        
    def learn(self,grid):
        self.learn_step_counter+=1
        if self.learn_step_counter>=self.learn_start and self.learn_step_counter%self.learn_frequent==0:
            print('start train',self.update_times)
            mini_batch = torch.cuda.FloatTensor(random.sample(self.replay_memory,self.batch_size))
            obs = mini_batch[:,0:8]
            obs[:,:4]=obs[:,:4]/420
            act = mini_batch[:,8].long()
            rew = mini_batch[:,9].unsqueeze(1)
            obs_ = mini_batch[:,10:-1]
            copy_obs_ = obs_[:,:2].clone().detach().cpu().data.numpy()
            obs_[:,:4]=obs_[:,:4]/420
            done = mini_batch[:,-1].unsqueeze(1)

            current_Q_values = self.evaluate_net(obs).gather(1, act.unsqueeze(1))
            #print(current_Q_values)
            #next_min_q = self.target_net(obs_).detach().min(1)[0]
            next_q = self.target_net(obs_).detach().cpu().data.numpy()
            next_min_q=[]
            for i in range(len(copy_obs_)):
                cur_posi = list(copy_obs_[i])
                node_ind = list(grid.keys())[list(grid.values()).index(cur_posi)]
                next_min_q.append(next_q[i][self.action_space[node_ind]].min())
            next_min_q = torch.cuda.FloatTensor(next_min_q)
            #print(next_max_q)
            next_Q_values = (1-done) * next_min_q.unsqueeze(1)
            #print(next_Q_values)
            target_Q_values = rew + (self.gamma * next_Q_values)
            #print(target_Q_values)
            loss = self.loss_func(current_Q_values, target_Q_values)
            #print(loss)
            '''print(obs)
            print(next_q)
            print(copy_obs_)
            print(next_min_q)
            print(done)
            print(next_Q_values)
            print(target_Q_values)'''
            '''current_Q_values = self.evaluate_net(obs).gather(1, act.unsqueeze(1))
            next_q = self.evaluate_net(obs_).detach().cpu().data.numpy()
            next_min_index=[]
            for i in range(len(copy_obs_)):
                cur_posi = list(copy_obs_[i])
                node_ind = list(grid.keys())[list(grid.values()).index(cur_posi)]
                next_min_index.append(np.where(next_q[i]==next_q[i][self.action_space[node_ind]].min())[0])
            
            next_min_index=torch.cuda.LongTensor(next_min_index)
            next_min_q = (self.target_net(obs_).detach()).gather(1, next_min_index)
            next_Q_values = (1-done) * next_min_q
            target_Q_values = rew + (self.gamma * next_Q_values)
            loss = self.loss_func(current_Q_values, target_Q_values)'''
            
            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()
            
            print('train finish')
            self.update_times += 1
            if self.update_times % self.target_replace == 0:
                self.target_net.load_state_dict(self.evaluate_net.state_dict())
                
            if self.update_times%50==0 and self.t>=self.t_min:
                self.t = self.t * self.decay
    
if __name__ == '__main__':
    model=DQN()
    #out = model.select_action_2(torch.tensor([[1,2,3,4,5,6,7,8],[8,7,6,5,4,3,2,1]],dtype=torch.float).cuda(),3)
    #print(out)
```
